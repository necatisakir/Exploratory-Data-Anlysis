```python
# eksik veriler; egitim butcesi ve giderleri ile yaslara gore ogrenci sayilari ve mat/okuma sinavlari not ortalamalari'na
# iliskin sutunlarda bulunmaktadir. yani PRIMARY_KEY, STATE ve YEAR sutunlari haric tum sutunlarda eksik veri var.

# bu kapsamda yil verisini kale almazsak:
  # ilgili sutunlara iliskin eksikleri; eyalet eyalet ayri ayri ortalama (mean) deger hesaplayarak doldurabiliriz.
  # ornegin/ CALIFORNIA'yi gruplayip sutunlarin ortalamasini aliriz ve eksik verileri bu degerlerle doldurabiliriz.
  # eyalet eyalet ayirdigimizda ise soyle bir sorun karsimiza cikiyor : bazi sutunlarda hicbir verinin olmadigi eyaletler var.
    # yani ornegin 'AS', 'BI', 'BIE' gbi eyalet satirlarinda  butce/finans verileri hic yok. dolayisiyla 
    # ortalama (mean) degeri cikmiyor
# yil verisini kale alip :
    # yil yil gruplayip her sutunun ortalamasini aldigimizda guzel bir iki yer haric mean sonuclari cikti. ama yil bazinda
      # ortalama almanin ve bu ortalamalari bos yerlere girmenin sikinti yaratabilecegini dusunuyorum. zira ornegin; 
      # CALIFORNIA ile UTAH in butcesi ve ogrenci sayilari cok farkli. CAL'daki bir eksik bilgi CAL ortalamasindan,
      # UTH'daki bir eksik bilgi UTH ortalamasindan alinsa daha anlamli olur bence. 
```

```python
## Niyetim, yil yil gruplarayarak her grubu kendi icinde sutun bazinda eksikleri tamamlamak.
  # Bunun icin once yil yil gruplayip ortalamma degerler hesaplanacak.
  # Daha sonra bu hesaplanan degerler her sutun icin ayri ayri eksik verilere doldurulacak.
```

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

```python
usegt1 = pd.read_csv('states_all.csv')
usegt1
```

```python
usegt1.info() # veri tipleri asagidadir. ilk iki sutun haric sayisaldir. 
```

```python
sutunlar = ['ENROLL','TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE', 
            'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE',
            'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G',
            'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 'GRADES_9_12_G', 
            'GRADES_ALL_G', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE',
            'AVG_READING_8_SCORE']
for sutun_adi in sutunlar:
    print("{} stunu iin problemli deerler : ".format(sutun_adi))
    for deger in usegt1[sutun_adi]:
        try:
            float(deger)
        except:
            print(deger) # toplam 1492 satirdan 'info' icerisine baktigimizda eksik olan sutunlarin hepsi 'float' formatinda
            # bunlardan eksik olanlarin da float formatinda oldugu asagida gordugumuz gibi teyit etmis olduk.
```

```python
usegt1.isnull().sum()*100/usegt1.shape[0] # eksik veriye sahip verilerin sutun basina orani asagidadir.
```

```python
usegt_yil = usegt1.groupby(['YEAR']).mean()
usegt_yil # yil yil gruplayarak her sutunun ortalaasini aldik. ancak eksik gorunen ortalama degerleri var.
```

```python
for i in usegt_yil.columns:
    usegt_yil[i].fillna(usegt_yil[i].mean(), inplace=True)
usegt_yil 

#eksik ortalama degerleri de kendi sutunu icinde bir daha ortalama alarak doldurduk. simdi elimizde her sutunun icerisinde
#bulunan eksik degerleri dolduracagimiz sayisal degerler mevcut. ornegin asagidaki tabloda 1995 yilinin ENROLL sutunundaki 
#eksik bilgileri "usegt_yil.at[1995, 'ENROLL']" ya da "usegt_yil.loc[1995, 'ENROLL']" kodunu kullanarak doldurabilirim.
```

```python
ilk_uc = usegt1[['PRIMARY_KEY', 'STATE', 'YEAR']] # eksik verisi bulunmayan sutunlari bir df'de birlesitirip daha sonra bunun
# yanina diger eksik sutunlari sirayla birlestiricem ve yeni temiz dosyam olusacak.
```

```python
usegt_yillar_top=pd.DataFrame() #yil yil gruplanmis satir bilgilerini alt alta ekleyerek nihai temiz df'e ulasmak icin
                                 # bos bir df tanimliyorum. asagidaki dongude her yil icin ilgili sutunlar yan yana 
                                 # birlestikten sonra, bu birlesen df'leri de alt alta birlestirerek nihai df'miz olusuyor.
for i in range(1992, 2018):
    top=ilk_uc[ilk_uc['YEAR']==i]
    for ii in usegt_yil.columns:
        zz=usegt1[usegt1['YEAR']==i]
        zz[ii].fillna(usegt_yil.at[i, ii], inplace=True) #
        top=pd.concat([top, zz[ii]], axis=1) # sutunlarin yanyana birlesmesi, ilave edilmesi     
    usegt_yillar_top=pd.concat([usegt_yillar_top, top]) # olusan yil gruplu satirlarin alt alta birlesmesi, ilave edilmesi
usegt_yillar_top
```

```python
usegt_yillar_top.isnull().sum()*100/usegt_yillar_top.shape[0] # vee sifir eksik dosyamiz
```
