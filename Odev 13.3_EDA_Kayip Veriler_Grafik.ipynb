```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from scipy import stats

warnings.filterwarnings('ignore')
```

```python
egttemiz = pd.read_csv('states_all_cleaned.csv', index_col=0)
egttemiz # bu temizlenmis dosyada kayip degerler, ortalamalar alinarak doldurulmustur.
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.hist(egttemiz['ENROLL'], bins=40, color='red')

plt.subplot(122)
stats.probplot(egttemiz['ENROLL'], plot=plt)

plt.show() #'ENROLL' sutununa ait verilerin normal dagilima uymadiklari acikca goruluyor. biraz saga yatik.
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.hist(egttemiz['TOTAL_REVENUE'], bins=40, color='red')

plt.subplot(122)
stats.probplot(egttemiz['TOTAL_REVENUE'], plot=plt)

plt.show() # Bu da saga yatik. normal bir dagilim sergilemiyor
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.hist(egttemiz['AVG_READING_8_SCORE'], bins=40, color='red')

plt.subplot(122)
stats.probplot(egttemiz['AVG_READING_8_SCORE'], plot=plt)

plt.show() # Normal dagilima bir pinc daha yakin. sanki.
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.hist(egttemiz['TOTAL_EXPENDITURE'], bins=40, color='red')

plt.subplot(122)
stats.probplot(egttemiz['TOTAL_EXPENDITURE'], plot=plt)

plt.show() # bu grafik de saga yatik. normal bir dagilim sergilemiyor.
```
