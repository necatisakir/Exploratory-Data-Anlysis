```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
usegt1 = pd.read_csv('states_all.csv')
usegt1
```

```python
usegt2=usegt1.interpolate() # df'mimizin orjinali sakata gelmesin diye enterpole edip yeni bir df'ye (usegt2) atiyoruz.
usegt2
```

```python
usegt2.isnull().sum()*100/usegt2.shape[0] #enterpole edilmis yeni usegt2 dosyamizin uc satirinda hala eksik bilgi var.
```

```python
usegtENROLL = usegt2[['ENROLL']]
usegtENROLL['ENROLL'].fillna(usegtENROLL['ENROLL'].mean(), inplace=True)
#usegtENROLL, eksik 'ENROLL' sutununun ortalamasini alaarak bos yerleri doldurduk
usegtOTEX = usegt2[['OTHER_EXPENDITURE']]
usegtOTEX['OTHER_EXPENDITURE'].fillna(usegtOTEX['OTHER_EXPENDITURE'].mean(), inplace=True)
#usegtOTEX, eksik 'OTHER EXPENDITURE' sutununun ortalamasini alaarak bos yerleri doldurduk
usegtRE8 = usegt2[['AVG_READING_8_SCORE']]
usegtRE8['AVG_READING_8_SCORE'].fillna(usegtRE8['AVG_READING_8_SCORE'].mean(), inplace=True)
#usegtRE8, eksik 'AVG_READING_8_SCORE' sutununun ortalamasini alaarak bos yerleri doldurduk
```

```python
usegt2[['ENROLL']] = usegtENROLL # enterpole edilmis df'nin ilgili sutunuyla yeni eksiksiz olan sutunu yer degistiriyoruz
usegt2[['OTHER_EXPENDITURE']] = usegtOTEX # yeni temiz sutunla yer degistiriyoruz
usegt2[['AVG_READING_8_SCORE']] = usegtRE8 # yeni temiz sutunla yer degistiriyoruz
```
